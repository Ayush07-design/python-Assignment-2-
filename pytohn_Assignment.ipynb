```python
## 1. What is Python, and why is it popular?
# Python is a high-level, interpreted language.
# **Popular because:**
# - Easy to read/write
# - Huge libraries (AI, web, data science)
# - Cross-platform
# - Big community

# ---

## 2. What is an interpreter in Python?
# Interpreter runs Python code **line-by-line**, translating it to machine code directly.

# ---

## 3. What are pre-defined keywords in Python?
# Reserved words that have special meaning.
# Examples: `if`, `else`, `while`, `for`, `def`

# ---

## 4. Can keywords be used as variable names?
#  No, keywords can’t be used as variable names.
# Example: `if = 10` ➝ Error

# ---

## 5. What is mutability in Python?
# - **Mutable** = can change (e.g., `list`)
# - **Immutable** = can't change (e.g., `tuple`, `int`)

# ---

## 6. Why are lists mutable, but tuples are immutable?
# - Lists allow modification (e.g., appending, removing)
# - Tuples are fixed for data safety & speed

# ---

## 7. Difference between `==` and `is`
# - `==` ➝ checks value
# - `is` ➝ checks object identity (memory address)

a = [1, 2]
b = [1, 2]
print(a == b)  # True
print(a is b)  # False

## 8. Logical operators
# and, or, not used in conditions

x = 10
y = 5
print(x > 5 and y < 10)  # True
print(x < 5 or y < 10)   # True
print(not(x > 5))        # False

## 9. What is type casting?
# Changing data type, e.g., str to int, int to float

## 10. Implicit vs Explicit Casting
# Implicit: auto convert → 5 + 2.5 → 7.5

# Explicit: manual → int("5"), float("3.14")

## 11. Purpose of conditional statements
# To make decisions using if, elif, else

## 12. How elif works
# elif means "else if"
# Used when first if is false.

x = 0
if x > 0:
    print("Positive")
elif x == 0:
    print("Zero")
else:
    print("Negative")

## 13. Difference between for and while loop
# for: use when you know how many times
# while: use when you repeat until a condition is false

## 14. Scenario for while loop
# Ask for correct password:

password = ""
while password != "secret":
    password = input("Enter password: ")

# Practical Python Programs

# 1. Print "Hello, World!"
print("Hello, World!")

# 2. Display name and age
name = "Aman"
age = 20
print("Name:", name)
print("Age:", age)

# 3. Print all Python keywords
import keyword
print(keyword.kwlist)

# 4. Check if a word is a Python keyword
import keyword

word = input("Enter a word: ")
if keyword.iskeyword(word):
    print(word, "is a Python keyword.")
else:
    print(word, "is not a Python keyword.")

# 5. List vs Tuple Mutability
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)

my_list[0] = 10  #  Works
print("Modified list:", my_list)

# my_tuple[0] = 10  #  Error if uncommented

# 6. Mutable vs Immutable in function
def modify_list(lst):
    lst[0] = 100

def modify_number(num):
    num = 100

a = [1, 2, 3]
b = 5

modify_list(a)
modify_number(b)

print("List after function:", a)
print("Number after function:", b)

# 7. Basic arithmetic
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Sum:", a + b)
print("Difference:", a - b)
print("Product:", a * b)
print("Quotient:", a / b)

# 8. Logical operators
x = 10
y = 5

print(x > 5 and y < 10)
print(x < 5 or y < 10)
print(not(x > 5))

# 9. Type casting user input
s = input("Enter a number: ")
i = int(s)
f = float(s)
b = bool(s)

print("Integer:", i)
print("Float:", f)
print("Boolean:", b)

# 10. Type casting list elements
str_list = ["1", "2", "3"]
int_list = [int(x) for x in str_list]
print(int_list)

# 11. Positive, Negative, or Zero
n = int(input("Enter a number: "))
if n > 0:
    print("Positive")
elif n < 0:
    print("Negative")
else:
    print("Zero")

# 12. For loop from 1 to 10
for i in range(1, 11):
    print(i)

# 13. Sum of even numbers 1 to 50
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i
print("Sum:", total)

# 14. Reverse a string using while loop
text = input("Enter a string: ")
i = len(text) - 1
reversed_str = ""

while i >= 0:
    reversed_str += text[i]
    i -= 1

print("Reversed:", reversed_str)

# 15. Factorial using while loop
num = int(input("Enter a number: "))
fact = 1

while num > 1:
    fact *= num
    num -= 1

print("Factorial:", fact)
```
